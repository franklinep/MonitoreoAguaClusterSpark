```python
import pandas as pd
import numpy as np
import unicodedata
from pathlib import Path

CSV_PATH = "/content/Monitoreo_Agua_Agua_superficial_1.csv"
OUT_DIR = Path("/content/json_out")
OUT_DIR.mkdir(parents=True, exist_ok=True)

def strip_accents(s: str) -> str:
    s = unicodedata.normalize("NFKD", s)
    return "".join(ch for ch in s if not unicodedata.combining(ch))

def normalize_col(name: str) -> str:
    name = strip_accents(name)
    name = name.strip().lower()
    name = name.replace(".", "_").replace("-", "_").replace(" ", "_")
    while "__" in name:
        name = name.replace("__", "_")
    name = "".join(ch for ch in name if (ch.isalnum() or ch == "_"))
    return name

def clean_str_series(s: pd.Series) -> pd.Series:
    s = s.astype("string")
    s = s.str.strip()
    s = s.replace({"-": pd.NA, "": pd.NA})
    return s

def to_float_series(s: pd.Series) -> pd.Series:
    s = s.astype("string")
    s = s.str.replace(",", ".", regex=False)
    s = s.str.replace(r"[^0-9\.\-]+", "", regex=True)
    return pd.to_numeric(s, errors="coerce")

df = pd.read_csv(CSV_PATH, low_memory=False)

df.columns = [normalize_col(c) for c in df.columns]

text_cols = [
    "txorigen","mes","expediente","cuc","coordinacion","txubigeo","punto_muestreo","tipo_punto",
    "laboratorio_anonimizado","tarea_tdr","matriz","parametro","signo_parametro","unidad_medida_parametro","rs"
]
for c in text_cols:
    if c in df.columns:
        df[c] = clean_str_series(df[c])

date_cols = ["fechaini","fechafin","fecha_pto"]
for c in date_cols:
    if c in df.columns:
        df[c] = pd.to_datetime(df[c], errors="coerce", dayfirst=True)

if "hora_pto" in df.columns:
    df["hora_pto"] = clean_str_series(df["hora_pto"])

numeric_cols = [
    "anho","txzona","coord_norte","coord_este",
    "otros","btex",
    "hidrocarburos_aromaticos_policiclicos_pah",
    "metales_disueltos",
    "metales_disueltos_por_icp_ms_incluido_hg",
    "metales_totales",
    "metales_totales_por_icp_ms_incluido_hg",
]
for c in numeric_cols:
    if c in df.columns:
        df[c] = to_float_series(df[c])

if "anho" in df.columns:
    df["anho"] = df["anho"].astype("Int64")

if "txzona" in df.columns:
    df["txzona"] = df["txzona"].astype("Int64")


if "mes" in df.columns:
    mes_norm = df["mes"].astype("string").fillna(pd.NA)
    mes_norm = mes_norm.map(lambda x: strip_accents(x).upper() if pd.notna(x) else x)

    mes_map = {
        "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
        "JULIO": 7, "AGOSTO": 8, "SETIEMBRE": 9, "SEPTIEMBRE": 9,
        "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
    }
    df["mes_num"] = mes_norm.map(mes_map).astype("Int64")


value_cols = [
    "otros","btex",
    "hidrocarburos_aromaticos_policiclicos_pah",
    "metales_disueltos",
    "metales_disueltos_por_icp_ms_incluido_hg",
    "metales_totales",
    "metales_totales_por_icp_ms_incluido_hg",
]
value_cols = [c for c in value_cols if c in df.columns]

arr = df[value_cols].to_numpy(dtype="float64")  # NaN donde no hay dato
mask = ~np.isnan(arr)
has_val = mask.any(axis=1)

first_idx = mask.argmax(axis=1)  # si no hay ninguno, devuelve 0 (lo corregimos con has_val)
tipo_muestra = np.array(value_cols, dtype="object")[first_idx]
resultado = arr[np.arange(len(df)), first_idx]

tipo_muestra = pd.Series(tipo_muestra, index=df.index, dtype="string")
resultado = pd.Series(resultado, index=df.index)

tipo_muestra = tipo_muestra.where(has_val, pd.NA)
resultado = resultado.where(has_val, np.nan)

df["tipo_muestra"] = tipo_muestra
df["resultado"] = resultado

dfm = df[df["resultado"].notna()].copy()

point_key_cols = ["punto_muestreo","tipo_punto","txubigeo","txzona","coord_norte","coord_este"]
point_key_cols = [c for c in point_key_cols if c in dfm.columns]

dim_puntos = dfm[point_key_cols].drop_duplicates().reset_index(drop=True)
dim_puntos["point_id"] = np.arange(1, len(dim_puntos) + 1, dtype=np.int64)

dfm = dfm.merge(dim_puntos, on=point_key_cols, how="left")


dfm["campaign_id"] = dfm["cuc"].astype("string")

camp_cols = [
    "campaign_id","cuc","expediente","txorigen","coordinacion",
    "fechaini","fechafin","anho","mes","mes_num",
    "tarea_tdr","matriz","laboratorio_anonimizado","rs"
]
camp_cols = [c for c in camp_cols if c in dfm.columns]

dim_campanas = dfm[camp_cols].drop_duplicates(subset=["campaign_id"]).reset_index(drop=True)


fact_cols = [
    "campaign_id","point_id","fecha_pto","hora_pto",
    "parametro","signo_parametro","unidad_medida_parametro",
    "tipo_muestra","resultado"
]
fact_cols = [c for c in fact_cols if c in dfm.columns]

fact_mediciones = dfm[fact_cols].copy()
fact_mediciones = fact_mediciones.reset_index(drop=True)
fact_mediciones["measurement_id"] = np.arange(1, len(fact_mediciones) + 1, dtype=np.int64)

dim_puntos_out = OUT_DIR / "dim_puntos.json"
dim_campanas_out = OUT_DIR / "dim_campanas.json"
fact_mediciones_out = OUT_DIR / "fact_mediciones.json"

dim_puntos.to_json(dim_puntos_out, orient="records", lines=True, force_ascii=False, date_format="iso")
dim_campanas.to_json(dim_campanas_out, orient="records", lines=True, force_ascii=False, date_format="iso")
fact_mediciones.to_json(fact_mediciones_out, orient="records", lines=True, force_ascii=False, date_format="iso")

print("JSON generados en:", OUT_DIR)
print("dim_puntos:", dim_puntos_out)
print("dim_campanas:", dim_campanas_out)
print("fact_mediciones:", fact_mediciones_out)
```
